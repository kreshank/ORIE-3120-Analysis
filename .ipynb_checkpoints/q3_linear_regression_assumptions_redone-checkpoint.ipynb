# Q3 Linear Regression: Noise Complaint Resolution Delay

This notebook builds a memory-safe linear regression workflow for Q3:

**How long does it take noise complaints to be resolved, and do some regions resolve them faster?**

The CSV is large, so the notebook:
- reads the file in chunks,
- keeps only relevant columns,
- samples filtered noise complaints,
- builds row-level transformed features,
- fits an OLS model on `log1p(delay_hours)`,
- checks linear regression assumptions using residual diagnostics.

The main editable section is the **feature map** cell. You should mostly only change that cell.

```python
from pathlib import Path
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

from scipy import stats
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson
from statsmodels.nonparametric.smoothers_lowess import lowess

warnings.filterwarnings("ignore", category=UserWarning)
pd.set_option("display.max_columns", 120)
pd.set_option("display.width", 160)
```

## 3. Column choices

These names are based on your sample output. The code below only uses columns that actually exist in the CSV.

```python
COL_CREATED = "Created Date"
COL_CLOSED = "Closed Date"
COL_DUE = "Due Date"
COL_RESOLUTION_UPDATED = "Resolution Action Updated Date"

COL_PROBLEM = "Problem (formerly Complaint Type)"
COL_DETAIL = "Problem Detail (formerly Descriptor)"
COL_ZIP = "Incident Zip"
COL_BOROUGH = "Borough"
COL_AGENCY = "Agency"
COL_STATUS = "Status"
COL_ADDRESS = "Incident Address"
COL_LAT = "Latitude"
COL_LON = "Longitude"

candidate_cols = [
    COL_CREATED, COL_CLOSED, COL_DUE, COL_RESOLUTION_UPDATED,
    COL_PROBLEM, COL_DETAIL, COL_ZIP, COL_BOROUGH, COL_AGENCY, COL_STATUS,
    COL_ADDRESS, COL_LAT, COL_LON
]

usecols = [c for c in candidate_cols if c in header]
missing_cols = [c for c in candidate_cols if c not in header]

print("Using columns:")
for c in usecols:
    print("  ", c)

print("\nMissing optional columns:")
for c in missing_cols:
    print("  ", c)
```

## 4. Memory-safe chunked loading

This cell filters to noise complaints and closed rows, computes `delay_hours`, and keeps a capped sample.

Important: this does **not** load the 4.5GB CSV into memory.

```python
def parse_311_datetime(s):
    # Your sample format: '01/01/2026 12:00:00 AM'
    # Let pandas infer if a few rows differ.
    return pd.to_datetime(s, errors="coerce", format="%m/%d/%Y %I:%M:%S %p")

def clean_zip(s):
    # Converts 10466.0 -> '10466', preserves missing as 'Missing'
    out = pd.to_numeric(s, errors="coerce").astype("Int64").astype("string")
    return out.fillna("Missing")

def prepare_chunk(chunk):
    chunk = chunk.copy()

    chunk["created_dt"] = parse_311_datetime(chunk[COL_CREATED])
    chunk["closed_dt"] = parse_311_datetime(chunk[COL_CLOSED])

    problem = chunk[COL_PROBLEM].astype("string").fillna("")
    is_noise = problem.str.contains("noise", case=False, na=False)

    status_ok = True
    if COL_STATUS in chunk.columns:
        status_ok = chunk[COL_STATUS].astype("string").str.lower().eq("closed")

    out = chunk.loc[is_noise & status_ok].copy()
    out["delay_hours"] = (out["closed_dt"] - out["created_dt"]).dt.total_seconds() / 3600

    out = out[
        out["created_dt"].notna()
        & out["closed_dt"].notna()
        & np.isfinite(out["delay_hours"])
        & (out["delay_hours"] >= 0)
    ].copy()

    if COL_ZIP in out.columns:
        out[COL_ZIP] = clean_zip(out[COL_ZIP])

    return out

rng = np.random.default_rng(RANDOM_SEED)
pieces = []
kept = 0
chunks_seen = 0
rows_seen = 0

reader = pd.read_csv(
    DATA_PATH,
    usecols=usecols,
    chunksize=CHUNKSIZE,
    low_memory=False,
)

for chunk in reader:
    chunks_seen += 1
    rows_seen += len(chunk)

    filtered = prepare_chunk(chunk)

    if len(filtered) > 0:
        remaining = MAX_FILTERED_ROWS - kept
        if remaining <= 0:
            break

        # If the chunk has more usable rows than we need, sample from it.
        if len(filtered) > remaining:
            filtered = filtered.sample(n=remaining, random_state=RANDOM_SEED + chunks_seen)

        pieces.append(filtered)
        kept += len(filtered)

    if chunks_seen % 10 == 0:
        print(f"chunks_seen={chunks_seen}, rows_seen={rows_seen:,}, kept_noise_rows={kept:,}")

    if kept >= MAX_FILTERED_ROWS:
        break

    if MAX_CHUNKS is not None and chunks_seen >= MAX_CHUNKS:
        break

df_raw = pd.concat(pieces, ignore_index=True) if pieces else pd.DataFrame()

print(f"chunks_seen: {chunks_seen}")
print(f"rows_seen: {rows_seen:,}")
print(f"filtered noise rows kept: {len(df_raw):,}")
df_raw.head()
```

## 5. Basic target cleanup

We model `log1p(delay_hours)` because resolution time is strongly right-skewed. This is usually better for linear regression than raw hours.

```python
df_model = df_raw.copy()

# Remove extreme tails for a cleaner linear regression assumption check.
# You can set this to 1.00 if you want no upper-tail cap.
TARGET_UPPER_QUANTILE = 0.95

upper = df_model["delay_hours"].quantile(TARGET_UPPER_QUANTILE)
df_model = df_model[df_model["delay_hours"] <= upper].copy()

df_model["log_delay"] = np.log1p(df_model["delay_hours"])

print("Rows after target cap:", len(df_model))
print("95th percentile delay cap:", upper)
display(df_model[["created_dt", "closed_dt", "delay_hours", "log_delay", COL_PROBLEM, COL_ZIP, COL_BOROUGH]].head())
df_model["delay_hours"].describe()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df_model["delay_hours"], bins=60)
plt.xlabel("Delay hours")
plt.ylabel("Count")
plt.title("Noise complaint resolution time, capped at selected quantile")
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(df_model["log_delay"], bins=60)
plt.xlabel("log1p(delay_hours)")
plt.ylabel("Count")
plt.title("Model target distribution")
plt.show()
```

## 6. Feature engineering functions

These functions create a copy of the dataframe and add transformed columns.

The idea is:

1. `add_engineered_columns(df)` creates many possible feature columns.
2. The feature map chooses which columns to include.
3. You only edit the feature map, not the model code.

```python
def clean_cat(s):
    return s.astype("string").fillna("Missing").astype(str)

def top_or_other(s, min_count=250):
    s = clean_cat(s)
    counts = s.value_counts(dropna=False)
    keep = counts[counts >= min_count].index
    return s.where(s.isin(keep), "Other")

def hour_float(dt):
    return dt.dt.hour + dt.dt.minute / 60 + dt.dt.second / 3600

def add_group_past_count(df, group_col, days, out_col):
    # Count previous complaints in the same group over the previous `days`.
    # This uses only the loaded sample, not the entire 4.5GB dataset.
    out = pd.Series(0.0, index=df.index)

    if group_col not in df.columns:
        df[out_col] = out
        return df

    tmp = df[["created_dt", group_col]].dropna(subset=["created_dt"]).copy()
    tmp["_group"] = clean_cat(tmp[group_col])
    window_ns = pd.Timedelta(days=days).value

    for _, g in tmp.groupby("_group", sort=False):
        g = g.sort_values("created_dt")
        times = g["created_dt"].astype("int64").to_numpy()
        left = np.searchsorted(times, times - window_ns, side="left")
        counts = np.arange(len(times)) - left
        out.loc[g.index] = counts

    df[out_col] = out
    return df

def add_engineered_columns(df):
    d = df.copy()

    # Basic categorical cleanup.
    if COL_PROBLEM in d.columns:
        d["problem_clean"] = clean_cat(d[COL_PROBLEM])
    if COL_DETAIL in d.columns:
        d["detail_clean"] = clean_cat(d[COL_DETAIL])
    if COL_ZIP in d.columns:
        d["zip_clean"] = clean_cat(d[COL_ZIP])
    if COL_BOROUGH in d.columns:
        d["borough_clean"] = clean_cat(d[COL_BOROUGH])
    if COL_AGENCY in d.columns:
        d["agency_clean"] = clean_cat(d[COL_AGENCY])
    if COL_ADDRESS in d.columns:
        d["address_clean"] = clean_cat(d[COL_ADDRESS])

    # Calendar features.
    d["hour"] = hour_float(d["created_dt"])
    d["hour_sq"] = d["hour"] ** 2
    d["day_of_week"] = d["created_dt"].dt.dayofweek.astype(float)
    d["month"] = d["created_dt"].dt.month.astype(float)
    d["is_weekend"] = d["created_dt"].dt.dayofweek.isin([5, 6]).astype(float)
    d["is_late_night"] = d["created_dt"].dt.hour.between(0, 5).astype(float)
    d["is_business_hours"] = (
        d["created_dt"].dt.dayofweek.between(0, 4)
        & d["created_dt"].dt.hour.between(9, 16)
    ).astype(float)

    # Cyclic encodings. Usually better than raw hour/day/month for cyclic patterns.
    d["hour_sin"] = np.sin(2 * np.pi * d["hour"] / 24)
    d["hour_cos"] = np.cos(2 * np.pi * d["hour"] / 24)

    d["dow_sin"] = np.sin(2 * np.pi * d["day_of_week"] / 7)
    d["dow_cos"] = np.cos(2 * np.pi * d["day_of_week"] / 7)

    d["month_sin"] = np.sin(2 * np.pi * d["month"] / 12)
    d["month_cos"] = np.cos(2 * np.pi * d["month"] / 12)

    # Time buckets as interpretable categorical features.
    d["hour_bucket"] = pd.cut(
        d["created_dt"].dt.hour,
        bins=[-1, 5, 11, 16, 20, 23],
        labels=["late_night", "morning", "afternoon", "evening", "night"]
    ).astype(str)

    # Operational-load features based on the loaded sample.
    if COL_ZIP in d.columns:
        d = add_group_past_count(d, COL_ZIP, days=7, out_col="zip_past_7d_count")
        d["log_zip_past_7d_count"] = np.log1p(d["zip_past_7d_count"])

    if COL_ADDRESS in d.columns:
        d = add_group_past_count(d, COL_ADDRESS, days=30, out_col="address_past_30d_count")
        d["log_address_past_30d_count"] = np.log1p(d["address_past_30d_count"])
        d["repeat_address_30d"] = (d["address_past_30d_count"] > 0).astype(float)

    # Interactions as categorical columns.
    if "zip_clean" in d.columns and "problem_clean" in d.columns:
        d["zip_x_problem"] = d["zip_clean"] + "__" + d["problem_clean"]

    if "borough_clean" in d.columns and "hour_bucket" in d.columns:
        d["borough_x_hour_bucket"] = d["borough_clean"] + "__" + d["hour_bucket"]

    return d

feature_df = add_engineered_columns(df_model)

print(feature_df.shape)
display(feature_df.head())
```

## 7. Feature map

Edit this cell when you want to change the regression.

### How to use this feature map

- `numeric`: continuous/binary columns included directly.
- `categorical`: columns one-hot encoded with rare levels collapsed to `"Other"`.
- `min_count`: minimum category frequency before a level becomes `"Other"`.

Examples:
- To add raw hour, include `"hour"` in `numeric`.
- To use cyclic hour, include both `"hour_sin"` and `"hour_cos"`.
- It is allowed to include both raw `"hour"` and `"hour_sin"/"hour_cos"`, but that can add multicollinearity. For interpretation, start with cyclic encodings; for prediction experiments, try both and compare train/test R².
- To add a new nonlinear feature, create it in `add_engineered_columns`, then add its column name here.

```python
FEATURE_MAP = {
    "numeric": [
        # Time features
        "hour",
        "hour_sq",
        "hour_sin",
        "hour_cos",
        "dow_sin",
        "dow_cos",
        "month_sin",
        "month_cos",
        "is_weekend",
        "is_late_night",
        "is_business_hours",

        # Operational-load features
        "log_zip_past_7d_count",
        "log_address_past_30d_count",
        "repeat_address_30d",
    ],

    "categorical": {
        "problem_clean": {"min_count": 250},
        "detail_clean": {"min_count": 300},
        "zip_clean": {"min_count": 350},
        "borough_clean": {"min_count": 1},
        "agency_clean": {"min_count": 1},

        # Interactions. These can improve fit, but can also add many columns.
        "zip_x_problem": {"min_count": 500},
        "borough_x_hour_bucket": {"min_count": 250},
    }
}
```

## 8. Build the design matrix

This cell turns the selected feature columns into a regression matrix.

```python
def build_design_matrix(df, feature_map):
    d = df.copy()

    numeric_cols = [c for c in feature_map.get("numeric", []) if c in d.columns]
    categorical_specs = {
        c: spec for c, spec in feature_map.get("categorical", {}).items()
        if c in d.columns
    }

    X_num = pd.DataFrame(index=d.index)
    for c in numeric_cols:
        X_num[c] = pd.to_numeric(d[c], errors="coerce").replace([np.inf, -np.inf], np.nan).fillna(0.0)

    cat_frames = []
    category_info = {}

    for c, spec in categorical_specs.items():
        min_count = spec.get("min_count", 250)
        collapsed = top_or_other(d[c], min_count=min_count)
        category_info[c] = collapsed.value_counts().head(20)
        cat_frames.append(pd.get_dummies(collapsed, prefix=c, drop_first=True, dtype=float))

    X_cat = pd.concat(cat_frames, axis=1) if cat_frames else pd.DataFrame(index=d.index)
    X = pd.concat([X_num, X_cat], axis=1)
    X = sm.add_constant(X, has_constant="add")

    return X, category_info

# Optional row cap for OLS stability.
reg_df = feature_df.copy()
if len(reg_df) > MAX_MODEL_ROWS:
    reg_df = reg_df.sample(n=MAX_MODEL_ROWS, random_state=RANDOM_SEED).sort_values("created_dt").copy()

X, category_info = build_design_matrix(reg_df, FEATURE_MAP)
y = reg_df["log_delay"].astype(float)

print("Rows:", len(reg_df))
print("Columns in X:", X.shape[1])
print("Numeric features used:", [c for c in FEATURE_MAP["numeric"] if c in reg_df.columns])
print("\nCategorical features used:")
for c in FEATURE_MAP["categorical"]:
    if c in reg_df.columns:
        print(" ", c)

display(X.head())
```

```python
print("Top retained category values after rare-level handling:")
for c, counts in category_info.items():
    print("\n" + c)
    display(counts)
```

## 9. Time-based train/test split

This is better than a random split here because 311 data is time-indexed. The model trains on earlier complaints and tests on later complaints.

```python
reg_df_sorted = reg_df.sort_values("created_dt").copy()
X_sorted = X.loc[reg_df_sorted.index]
y_sorted = y.loc[reg_df_sorted.index]

split = int(0.8 * len(reg_df_sorted))

X_train = X_sorted.iloc[:split]
X_test = X_sorted.iloc[split:]
y_train = y_sorted.iloc[:split]
y_test = y_sorted.iloc[split:]

df_train = reg_df_sorted.iloc[:split]
df_test = reg_df_sorted.iloc[split:]

print("Train rows:", len(X_train))
print("Test rows:", len(X_test))
print("Train date range:", df_train["created_dt"].min(), "to", df_train["created_dt"].max())
print("Test date range:", df_test["created_dt"].min(), "to", df_test["created_dt"].max())
```

## 10. Fit OLS with robust standard errors

OLS itself is CPU-based here. The GPU is not very useful for this particular assumption-testing workflow because `statsmodels` runs on CPU. A GPU would matter more for large-scale prediction models such as XGBoost, LightGBM, RAPIDS/cuML, or neural nets, but those are less direct for linear-regression assumption checks.

```python
model = sm.OLS(y_train, X_train).fit(cov_type="HC3")

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

def eval_regression(y_true, y_pred, label):
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{label} RMSE, log scale: {rmse:.4f}")
    print(f"{label} MAE, log scale:  {mae:.4f}")
    print(f"{label} R^2:             {r2:.4f}")

eval_regression(y_train, pred_train, "Train")
print()
eval_regression(y_test, pred_test, "Test")

print("\nStatsmodels in-sample R^2:", model.rsquared)
print("Statsmodels adj. R^2:", model.rsquared_adj)
```

```python
# Full summary can be very long when there are many dummies.
# Showing the coefficient table is useful, but avoid printing thousands of rows.
summary_table = model.summary2().tables[1].copy()
summary_table["abs_t"] = summary_table["t"].abs()
display(summary_table.sort_values("abs_t", ascending=False).head(40))

print(model.summary())
```

## 11. Residual diagnostics for linear regression assumptions

These plots/checks focus on:
- linearity,
- constant variance,
- approximate normality of residuals,
- independence/autocorrelation,
- influential points.

With 311 data, assumptions will probably not be perfect. The goal is to identify the main violations and improve the model.

```python
resid = y_train - pred_train
fitted = pred_train

diagnostics_df = df_train.copy()
diagnostics_df["fitted"] = fitted
diagnostics_df["resid"] = resid
diagnostics_df["abs_resid"] = np.abs(resid)
diagnostics_df["sqrt_abs_standardized_resid"] = np.sqrt(np.abs((resid - resid.mean()) / resid.std()))

print(resid.describe())
```

### 11.1 Linearity: residuals vs fitted

A good pattern is residuals randomly scattered around zero. A strong curve suggests nonlinear structure is missing.

```python
plot_sample = diagnostics_df.sample(min(len(diagnostics_df), 8000), random_state=RANDOM_SEED)

plt.figure(figsize=(8, 5))
plt.scatter(plot_sample["fitted"], plot_sample["resid"], alpha=0.25, s=8)
smooth = lowess(plot_sample["resid"], plot_sample["fitted"], frac=0.25, return_sorted=True)
plt.plot(smooth[:, 0], smooth[:, 1], linewidth=2)
plt.axhline(0, linestyle="--")
plt.xlabel("Fitted log delay")
plt.ylabel("Residual")
plt.title("Residuals vs fitted")
plt.show()
```

### 11.2 Constant variance: scale-location plot and Breusch-Pagan test

If the smooth line trends upward/downward, variance changes with fitted values.

```python
plt.figure(figsize=(8, 5))
plt.scatter(plot_sample["fitted"], plot_sample["sqrt_abs_standardized_resid"], alpha=0.25, s=8)
smooth = lowess(plot_sample["sqrt_abs_standardized_resid"], plot_sample["fitted"], frac=0.25, return_sorted=True)
plt.plot(smooth[:, 0], smooth[:, 1], linewidth=2)
plt.xlabel("Fitted log delay")
plt.ylabel("sqrt(abs(standardized residual))")
plt.title("Scale-location plot")
plt.show()

bp_stat, bp_pvalue, f_stat, f_pvalue = het_breuschpagan(resid, X_train)
print("Breusch-Pagan LM statistic:", bp_stat)
print("Breusch-Pagan LM p-value:", bp_pvalue)
print("Breusch-Pagan F statistic:", f_stat)
print("Breusch-Pagan F p-value:", f_pvalue)
```

### 11.3 Normality: Q-Q plot

OLS coefficients do not require perfectly normal residuals for unbiasedness, but strong non-normality affects inference and prediction intervals.

```python
sm.qqplot(resid, line="45", fit=True)
plt.title("Q-Q plot of residuals")
plt.show()

jb_stat, jb_pvalue = stats.jarque_bera(resid)
print("Jarque-Bera statistic:", jb_stat)
print("Jarque-Bera p-value:", jb_pvalue)
```

### 11.4 Independence: residuals over time and Durbin-Watson

A visible time pattern means there may be temporal structure not captured by the features.

```python
time_plot = diagnostics_df.sort_values("created_dt").copy()
if len(time_plot) > 8000:
    time_plot = time_plot.iloc[np.linspace(0, len(time_plot)-1, 8000).astype(int)]

plt.figure(figsize=(10, 4))
plt.plot(time_plot["created_dt"], time_plot["resid"], linewidth=0.6)
plt.axhline(0, linestyle="--")
plt.xlabel("Created date")
plt.ylabel("Residual")
plt.title("Residuals over time")
plt.xticks(rotation=30)
plt.show()

print("Durbin-Watson:", durbin_watson(resid))
```

### 11.5 Influential points

This can be somewhat slow if the design matrix has many columns, so it runs on a smaller subsample.

```python
INFLUENCE_N = min(len(X_train), 8000)

if len(X_train) > INFLUENCE_N:
    idx = X_train.sample(n=INFLUENCE_N, random_state=RANDOM_SEED).index
    small_model = sm.OLS(y_train.loc[idx], X_train.loc[idx]).fit()
else:
    idx = X_train.index
    small_model = sm.OLS(y_train, X_train).fit()

influence = small_model.get_influence()
cooks = influence.cooks_distance[0]

plt.figure(figsize=(8, 4))
plt.scatter(np.arange(len(cooks)), cooks, s=8, alpha=0.4)
plt.axhline(4 / len(cooks), linestyle="--")
plt.xlabel("Observation index in influence sample")
plt.ylabel("Cook's distance")
plt.title("Cook's distance")
plt.show()

print("Cook's distance threshold 4/n:", 4 / len(cooks))
print("Number above threshold:", int((cooks > 4 / len(cooks)).sum()))
```

## 12. Feature-group comparison

This compares simple model variants. Use it to see whether the extra transformed features actually help test-set performance.

```python
FEATURE_VARIANTS = {
    "location_only": {
        "numeric": [],
        "categorical": {
            "zip_clean": {"min_count": 350},
            "borough_clean": {"min_count": 1},
        },
    },
    "complaint_location": {
        "numeric": [],
        "categorical": {
            "problem_clean": {"min_count": 250},
            "detail_clean": {"min_count": 300},
            "zip_clean": {"min_count": 350},
            "borough_clean": {"min_count": 1},
        },
    },
    "complaint_location_time": {
        "numeric": [
            "hour", "hour_sq", "hour_sin", "hour_cos",
            "dow_sin", "dow_cos", "month_sin", "month_cos",
            "is_weekend", "is_late_night", "is_business_hours",
        ],
        "categorical": {
            "problem_clean": {"min_count": 250},
            "detail_clean": {"min_count": 300},
            "zip_clean": {"min_count": 350},
            "borough_clean": {"min_count": 1},
        },
    },
    "full_feature_map": FEATURE_MAP,
}

rows = []

for name, fmap in FEATURE_VARIANTS.items():
    Xv, _ = build_design_matrix(reg_df, fmap)
    Xv_sorted = Xv.loc[reg_df_sorted.index]

    Xv_train = Xv_sorted.iloc[:split]
    Xv_test = Xv_sorted.iloc[split:]

    m = sm.OLS(y_train, Xv_train).fit(cov_type="HC3")
    tr = m.predict(Xv_train)
    te = m.predict(Xv_test)

    rows.append({
        "variant": name,
        "n_features": Xv_train.shape[1],
        "train_r2": r2_score(y_train, tr),
        "test_r2": r2_score(y_test, te),
        "test_rmse_log": mean_squared_error(y_test, te, squared=False),
        "test_mae_log": mean_absolute_error(y_test, te),
    })

variant_results = pd.DataFrame(rows).sort_values("test_r2", ascending=False)
display(variant_results)
```

## 13. Optional: ZIP-level residual view

This checks whether certain regions are still systematically under/over-predicted after controlling for the selected features.

```python
if COL_ZIP in diagnostics_df.columns:
    zip_resid = (
        diagnostics_df
        .assign(zip_clean=clean_cat(diagnostics_df[COL_ZIP]))
        .groupby("zip_clean")
        .agg(
            n=("resid", "size"),
            mean_resid=("resid", "mean"),
            mean_delay_hours=("delay_hours", "mean"),
        )
        .query("n >= 25")
        .sort_values("mean_resid", ascending=False)
    )

    print("ZIPs with largest positive residuals: actual delays longer than model predicts")
    display(zip_resid.head(15))

    print("ZIPs with largest negative residuals: actual delays shorter than model predicts")
    display(zip_resid.tail(15))

    plt.figure(figsize=(8, 5))
    plt.scatter(zip_resid["mean_delay_hours"], zip_resid["mean_resid"], s=np.sqrt(zip_resid["n"]) * 3, alpha=0.5)
    plt.axhline(0, linestyle="--")
    plt.xlabel("Average raw delay hours by ZIP")
    plt.ylabel("Average residual by ZIP")
    plt.title("ZIP-level residual pattern")
    plt.show()
else:
    print("No ZIP column available.")
```

## 14. Notes for iteration

Things to try in the feature map:

1. Start simple: complaint type/detail + ZIP + borough + cyclic time.
2. Add operational load features.
3. Add interactions only if test-set R² improves.
4. If OLS gets slow, increase `min_count` on categorical features or reduce `MAX_MODEL_ROWS`.
5. If residual-vs-fitted shows curvature, add nonlinear columns in `add_engineered_columns`, then select them in `FEATURE_MAP`.
6. If the Breusch-Pagan p-value is small, keep using robust standard errors (`cov_type="HC3"`).
7. Avoid using `Closed Date`, `Resolution Action Updated Date`, or any post-resolution column as a feature, because those leak the answer.
