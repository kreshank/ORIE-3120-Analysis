# Q3 Linear Regression Assumption Checks

Goal: model **noise complaint resolution delay** and check the main linear regression assumptions from lecture:

1. linearity / $E[\epsilon_i \mid X_i] = 0$
2. constant variance
3. independence
4. normality of residuals

This notebook is written to avoid loading the full 4.5GB CSV into memory. It reads only selected columns, filters to noise complaints, and samples rows chunk-by-chunk.

```python
from pathlib import Path
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statsmodels.nonparametric.smoothers_lowess import lowess
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson, jarque_bera

DATA_PATH = Path("data/311_data.csv")
DATE_FORMAT = "%m/%d/%Y %I:%M:%S %p"

# Main safety knobs. Lower these if Jupyter feels slow.
CHUNKSIZE = 200_000
MAX_MODEL_ROWS = 120_000
PER_CHUNK_MAX_ROWS = 3_000
PLOT_SAMPLE_ROWS = 8_000
LOWESS_ROWS = 4_000
RANDOM_STATE = 3120
```

## Feature config

Change features here only. The rest of the notebook reads this dictionary.

- Set `enabled` to `False` to drop a feature.
- For categorical variables, `max_levels` controls how many dummy variables are created. This prevents huge design matrices.
- Derived features use `kind = "derived"`.

```python
features = {
    "zip_code": {
        "enabled": True,
        "kind": "categorical",
        "source": "Incident Zip",
        "max_levels": 40,
    },
    "complaint_type": {
        "enabled": True,
        "kind": "categorical",
        "source": "Problem (formerly Complaint Type)",
        "max_levels": 12,
    },
    "complaint_detail": {
        "enabled": True,
        "kind": "categorical",
        "source": "Problem Detail (formerly Descriptor)",
        "max_levels": 25,
    },
    "borough": {
        "enabled": True,
        "kind": "categorical",
        "source": "Borough",
        "max_levels": 10,
    },
    "agency": {
        "enabled": True,
        "kind": "categorical",
        "source": "Agency",
        "max_levels": 12,
    },
    "day_of_week": {
        "enabled": True,
        "kind": "derived",
        "source": "Created Date",
        "max_levels": 7,
    },
    "time_bucket": {
        "enabled": True,
        "kind": "derived",
        "source": "Created Date",
        "max_levels": 5,
    },
    "month": {
        "enabled": True,
        "kind": "derived",
        "source": "Created Date",
        "max_levels": 12,
    },
    "repeat_address_flag": {
        "enabled": True,
        "kind": "derived",
        "source": "Incident Address",
    },
    "zip_volume_proxy": {
        "enabled": True,
        "kind": "derived_numeric",
        "source": "Incident Zip",
    },
}

TARGET_NAME = "delay_hours"
TARGET_TRANSFORM = "log1p"   # use "none" or "log1p"; log1p usually helps with right-skewed delays
DELAY_CAP_QUANTILE = 0.95     # cap extreme closed-case delays before fitting
KEEP_ONLY_NOISE = True
```

```python
def read_header(path):
    return list(pd.read_csv(path, nrows=0).columns)

all_columns = read_header(DATA_PATH)
print(f"Number of columns in CSV: {len(all_columns)}")

base_cols = ["Created Date", "Closed Date", "Status", "Problem (formerly Complaint Type)"]
feature_source_cols = [cfg.get("source") for cfg in features.values() if cfg.get("enabled")]
needed_cols = []
for c in base_cols + feature_source_cols:
    if c and c in all_columns and c not in needed_cols:
        needed_cols.append(c)

missing_cols = sorted(set(base_cols + feature_source_cols) - set(all_columns))
print("Reading these columns only:")
for c in needed_cols:
    print(" -", c)

if missing_cols:
    print("\nColumns requested but not found; those features will be skipped:")
    for c in missing_cols:
        print(" -", c)
```

```python
def parse_311_dates(s):
    return pd.to_datetime(s, format=DATE_FORMAT, errors="coerce")


def process_chunk(chunk):
    chunk = chunk.copy()

    # Q3 is about noise complaints.
    if KEEP_ONLY_NOISE and "Problem (formerly Complaint Type)" in chunk.columns:
        is_noise = chunk["Problem (formerly Complaint Type)"].astype(str).str.contains("Noise", case=False, na=False)
        chunk = chunk.loc[is_noise]

    if chunk.empty:
        return chunk

    created = parse_311_dates(chunk["Created Date"])
    closed = parse_311_dates(chunk["Closed Date"])
    chunk["created_dt"] = created
    chunk["closed_dt"] = closed
    chunk[TARGET_NAME] = (closed - created).dt.total_seconds() / 3600

    # For linear regression on time-to-resolution, keep cases with usable closed time.
    chunk = chunk.loc[chunk[TARGET_NAME].notna()]
    chunk = chunk.loc[chunk[TARGET_NAME] >= 0]
    chunk = chunk.loc[chunk["created_dt"].notna()]

    return chunk


def load_sampled_311(path):
    parts = []
    total_seen_after_filter = 0

    reader = pd.read_csv(
        path,
        usecols=needed_cols,
        chunksize=CHUNKSIZE,
        low_memory=False,
    )

    for k, chunk in enumerate(reader, start=1):
        chunk = process_chunk(chunk)
        n = len(chunk)
        total_seen_after_filter += n

        if n > 0:
            keep_n = min(n, PER_CHUNK_MAX_ROWS)
            sampled = chunk.sample(n=keep_n, random_state=RANDOM_STATE + k)
            parts.append(sampled)

        # Keep memory bounded even if many chunks are read.
        current_rows = sum(len(p) for p in parts)
        if current_rows > 2 * MAX_MODEL_ROWS:
            tmp = pd.concat(parts, ignore_index=True)
            tmp = tmp.sample(n=MAX_MODEL_ROWS, random_state=RANDOM_STATE + k)
            parts = [tmp]

        if k % 10 == 0:
            print(f"processed chunks: {k}, filtered closed noise rows seen: {total_seen_after_filter:,}, sample rows kept: {sum(len(p) for p in parts):,}")

    if not parts:
        raise ValueError("No usable rows found. Check the target/date columns and noise complaint filter.")

    df = pd.concat(parts, ignore_index=True)
    if len(df) > MAX_MODEL_ROWS:
        df = df.sample(n=MAX_MODEL_ROWS, random_state=RANDOM_STATE)

    return df.reset_index(drop=True), total_seen_after_filter


df_raw, total_seen = load_sampled_311(DATA_PATH)
print("\nFinal sampled shape:", df_raw.shape)
print("Total filtered closed noise rows seen while scanning:", f"{total_seen:,}")
print("Approx sample memory MB:", round(df_raw.memory_usage(deep=True).sum() / 1024**2, 2))
display(df_raw.head(3).T)
```

```python
# Basic target check before fitting.
cap_value = df_raw[TARGET_NAME].quantile(DELAY_CAP_QUANTILE)
df_model = df_raw.loc[df_raw[TARGET_NAME].between(0, cap_value)].copy()

if TARGET_TRANSFORM == "log1p":
    df_model["y"] = np.log1p(df_model[TARGET_NAME])
    y_label = "log(1 + delay_hours)"
elif TARGET_TRANSFORM == "none":
    df_model["y"] = df_model[TARGET_NAME]
    y_label = "delay_hours"
else:
    raise ValueError("TARGET_TRANSFORM must be 'none' or 'log1p'")

print("Rows before delay cap:", len(df_raw))
print("Rows after delay cap:", len(df_model))
print(f"Delay cap at q={DELAY_CAP_QUANTILE}:", round(cap_value, 3), "hours")

summary = df_model[TARGET_NAME].describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95])
display(summary.to_frame("delay_hours"))

plt.figure(figsize=(7, 4))
plt.hist(df_model[TARGET_NAME], bins=60)
plt.xlabel("delay_hours")
plt.ylabel("count")
plt.title("Noise complaint resolution delay after cap")
plt.show()
```

```python
def clean_category(s):
    return s.astype("string").fillna("Missing").str.strip().replace("", "Missing")


def top_level_other(s, max_levels):
    s = clean_category(s)
    top = s.value_counts().head(max_levels).index
    return s.where(s.isin(top), other="Other")


def add_derived_features(df, features):
    out = df.copy()

    if features.get("day_of_week", {}).get("enabled", False):
        out["feat__day_of_week"] = out["created_dt"].dt.day_name()

    if features.get("time_bucket", {}).get("enabled", False):
        hour = out["created_dt"].dt.hour
        out["feat__time_bucket"] = pd.cut(
            hour,
            bins=[-1, 5, 11, 16, 20, 23],
            labels=["late_night", "morning", "afternoon", "evening", "night"],
        ).astype("string")

    if features.get("month", {}).get("enabled", False):
        out["feat__month"] = out["created_dt"].dt.month.astype("Int64").astype("string")

    if features.get("repeat_address_flag", {}).get("enabled", False):
        col = features["repeat_address_flag"].get("source")
        if col in out.columns:
            address_counts = out[col].astype("string").fillna("Missing").value_counts()
            out["feat__repeat_address_flag"] = out[col].astype("string").map(address_counts).fillna(0).gt(1).astype(int)
        else:
            print("Skipping repeat_address_flag because source column is missing:", col)

    if features.get("zip_volume_proxy", {}).get("enabled", False):
        col = features["zip_volume_proxy"].get("source")
        if col in out.columns:
            z = clean_category(out[col])
            zip_counts = z.value_counts()
            out["feat__zip_volume_proxy"] = np.log1p(z.map(zip_counts).astype(float))
        else:
            print("Skipping zip_volume_proxy because source column is missing:", col)

    return out


def build_design_matrix(df, features):
    df = add_derived_features(df, features)
    pieces = []
    feature_groups = {}
    used_feature_names = []

    for name, cfg in features.items():
        if not cfg.get("enabled", False):
            continue

        kind = cfg.get("kind")
        source = cfg.get("source")
        max_levels = cfg.get("max_levels", 20)

        if kind == "categorical":
            if source not in df.columns:
                print(f"Skipping {name}: missing source column {source}")
                continue
            cat = top_level_other(df[source], max_levels=max_levels)
            dummies = pd.get_dummies(cat, prefix=name, drop_first=True, dtype=float)
            pieces.append(dummies)
            feature_groups[name] = list(dummies.columns)
            used_feature_names.append(name)

        elif kind == "derived":
            col = f"feat__{name}"
            if col not in df.columns:
                print(f"Skipping {name}: derived column was not created")
                continue
            cat = top_level_other(df[col], max_levels=max_levels)
            dummies = pd.get_dummies(cat, prefix=name, drop_first=True, dtype=float)
            pieces.append(dummies)
            feature_groups[name] = list(dummies.columns)
            used_feature_names.append(name)

        elif kind == "derived_numeric":
            col = f"feat__{name}"
            if col not in df.columns:
                print(f"Skipping {name}: derived numeric column was not created")
                continue
            numeric = pd.to_numeric(df[col], errors="coerce").fillna(0).astype(float).to_frame(name)
            pieces.append(numeric)
            feature_groups[name] = [name]
            used_feature_names.append(name)

    if not pieces:
        raise ValueError("No enabled features were available. Edit the features dictionary.")

    X = pd.concat(pieces, axis=1)
    X = sm.add_constant(X, has_constant="add")
    y = df["y"].astype(float)
    return X.astype(float), y, df, feature_groups, used_feature_names


X, y, model_df, feature_groups, used_feature_names = build_design_matrix(df_model, features)
print("X shape:", X.shape)
print("Used feature groups:", used_feature_names)
display(X.head())
```

```python
model = sm.OLS(y, X).fit()
print(model.summary())

model_df = model_df.copy()
model_df["fitted"] = model.fittedvalues
model_df["resid"] = model.resid
model_df["std_resid"] = model.get_influence().resid_studentized_internal

print("R^2:", round(model.rsquared, 4))
print("Adjusted R^2:", round(model.rsquared_adj, 4))
```

## Rough variance decomposition by feature group

This is not a perfect causal decomposition. It asks: how much does model $R^2$ drop when each feature group is removed from the fitted design matrix?

```python
def r2_after_dropping_group(X, y, cols_to_drop, max_rows=50_000):
    if len(X) > max_rows:
        idx = np.random.default_rng(RANDOM_STATE).choice(len(X), size=max_rows, replace=False)
        X_use = X.iloc[idx]
        y_use = y.iloc[idx]
    else:
        X_use = X
        y_use = y

    keep_cols = [c for c in X_use.columns if c not in cols_to_drop]
    reduced = sm.OLS(y_use, X_use[keep_cols]).fit()
    full = sm.OLS(y_use, X_use).fit()
    return full.rsquared, reduced.rsquared

rows = []
for group, cols in feature_groups.items():
    full_r2, reduced_r2 = r2_after_dropping_group(X, y, cols)
    rows.append({
        "feature_group": group,
        "full_R2_on_same_sample": full_r2,
        "R2_without_group": reduced_r2,
        "delta_R2": full_r2 - reduced_r2,
        "num_columns": len(cols),
    })

r2_decomp = pd.DataFrame(rows).sort_values("delta_R2", ascending=False)
display(r2_decomp)
```

## Assumption 1: linearity / $E[\epsilon_i \mid X_i] = 0$

The residuals should be centered around 0 with no systematic pattern. The LOWESS curve should be close to horizontal at 0.

```python
plot_df = model_df.sample(n=min(PLOT_SAMPLE_ROWS, len(model_df)), random_state=RANDOM_STATE).copy()
lowess_df = plot_df.sample(n=min(LOWESS_ROWS, len(plot_df)), random_state=RANDOM_STATE).sort_values("fitted")
smoothed = lowess(lowess_df["resid"], lowess_df["fitted"], frac=0.30, return_sorted=True)

plt.figure(figsize=(7, 4))
plt.scatter(plot_df["fitted"], plot_df["resid"], s=4, alpha=0.25)
plt.plot(smoothed[:, 0], smoothed[:, 1], linewidth=2)
plt.axhline(0, linestyle="--", linewidth=1)
plt.xlabel("fitted values")
plt.ylabel("residuals")
plt.title("Residuals vs fitted values with LOWESS")
plt.show()
```

```python
# Residual means by selected categorical features. Large deviations from 0 suggest remaining structure.
for group in ["borough", "complaint_type", "zip_code", "time_bucket", "day_of_week"]:
    cfg = features.get(group, {})
    if not cfg.get("enabled", False):
        continue

    if cfg.get("kind") == "categorical":
        col = cfg.get("source")
    else:
        col = f"feat__{group}"

    if col not in model_df.columns:
        continue

    tmp = model_df.copy()
    tmp[col] = top_level_other(tmp[col], max_levels=cfg.get("max_levels", 15))
    by = tmp.groupby(col, dropna=False).agg(
        mean_resid=("resid", "mean"),
        n=("resid", "size"),
        mean_delay_hours=(TARGET_NAME, "mean"),
    ).sort_values("n", ascending=False).head(15)

    print("\n", group)
    display(by)
```

## Assumption 2: constant variance

Residual spread should be roughly stable across fitted values. The Breusch-Pagan test gives a formal check, but with large samples it can flag even small deviations.

```python
plt.figure(figsize=(7, 4))
plt.scatter(plot_df["fitted"], np.sqrt(np.abs(plot_df["std_resid"])), s=4, alpha=0.25)
plt.xlabel("fitted values")
plt.ylabel("sqrt(abs(standardized residual))")
plt.title("Scale-location plot")
plt.show()

bp_n = min(50_000, len(model_df))
bp_idx = np.random.default_rng(RANDOM_STATE).choice(len(model_df), size=bp_n, replace=False)
bp_resid = model_df["resid"].iloc[bp_idx]
bp_exog = X.iloc[bp_idx]

bp_stat, bp_pvalue, f_stat, f_pvalue = het_breuschpagan(bp_resid, bp_exog)
print("Breusch-Pagan LM statistic:", bp_stat)
print("Breusch-Pagan p-value:", bp_pvalue)
print("F statistic:", f_stat)
print("F-test p-value:", f_pvalue)
```

## Assumption 3: independence

For 311 complaints, independence is not guaranteed because nearby complaints or complaints close in time may be related. As a basic check, sort residuals by created time and look for autocorrelation.

```python
time_df = model_df.sort_values("created_dt").copy()

dw = durbin_watson(time_df["resid"])
print("Durbin-Watson statistic after sorting by created time:", dw)
print("Rule of thumb: around 2 means little first-order autocorrelation; below 2 suggests positive autocorrelation.")

line_df = time_df.sample(n=min(PLOT_SAMPLE_ROWS, len(time_df)), random_state=RANDOM_STATE).sort_values("created_dt")
line_df["rolling_resid_mean"] = line_df["resid"].rolling(200, min_periods=30).mean()

plt.figure(figsize=(8, 4))
plt.plot(line_df["created_dt"], line_df["rolling_resid_mean"])
plt.axhline(0, linestyle="--", linewidth=1)
plt.xlabel("created time")
plt.ylabel("rolling mean residual")
plt.title("Residual drift over time")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## Assumption 4: normality

Normal residuals mainly matter for exact small-sample inference. With many rows, predictions and coefficient estimates can still be useful even if residuals are not perfectly normal.

```python
qq_resid = plot_df["std_resid"].replace([np.inf, -np.inf], np.nan).dropna()

plt.figure(figsize=(5, 5))
sm.qqplot(qq_resid, line="45", fit=True, ax=plt.gca())
plt.title("Q-Q plot of standardized residuals")
plt.show()

plt.figure(figsize=(7, 4))
plt.hist(qq_resid, bins=60)
plt.xlabel("standardized residual")
plt.ylabel("count")
plt.title("Histogram of standardized residuals")
plt.show()

jb_stat, jb_pvalue, skew, kurtosis = jarque_bera(qq_resid)
print("Jarque-Bera statistic:", jb_stat)
print("Jarque-Bera p-value:", jb_pvalue)
print("skew:", skew)
print("kurtosis:", kurtosis)
```
